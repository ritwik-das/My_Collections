### <span style="color:#BB1C75">Linear Regression</span>

This helps to model the relationship between the dependent variable and independent variables as a straight line. The equation is given by, 

$$\hat{y}(w,x)=b+w_1x_1+w_2x_2+...+w_nx_n \equiv b+w^Tx$$ 

where,
<ul>
    <li>$N$ : number of data points</li>
    <li>$d$ : number of features</li>
    <li>$k$ : number of classes</li>
    <li>$w$ : weight vector ($d\times k$) ; each column = weights for one class</li>
    <li>$x$ : input feature matrix ($N\times d$) ; each row = one sample, each column = one feature</li>
    <li>$b$ : intercept or bias ($1\times k$) ; bias term for each class</li>
    <li>$\hat{y}$ : predicted value ($N\times k$) ; each row = output per one sample, each column = output per one class</li>
</ul>

The common cost function used for linear regression is Mean-square-Error(MSE), given by, $$\mathcal{L}(b,w_0,w_1,...,w_n) = \frac{1}{2N} \sum_{i=1}^{N}\;(y_i-\hat{y}_i)^2 \equiv \frac{1}{2N} \sum_{i=1}^{N}\;\{y_i-(b+w^Tx_i)\}^2$$ where, $y$ is the target data that comes from observations or training dataset and $(y_i-\hat{y}_i)$ is the loss function for one sample.

The aim is to minimize the cost function ($\mathcal{L}$) in terms of the residual between the predicted output ($\hat{y}$) and the target value ($y$).

Generally, there are two way to estimate the model parameters : 
<ol>
    <li>Ordinary Least Squares (OLS)</li>
    <li>Stochastic Gradient Descent (SGD)</li>
</ol>


<b>Ordinary Least Squares (OLS)</b>

The solution is obtained analytically by setting the gradient to zero : $\frac{\partial\mathcal{L}}{\partial w} = 0$ that leads $w = (x^Tx)^{-1}x^Ty$. This is the closed-form solution but this method is slower for large datasets, because computing $(x^Tx)$ and matrix inversion is expensive for very high dimensions.

<b>Stochastic Gradient Descent (SGD)</b>

This minimize the same cost function, but in an iterative manner : $\theta =: \theta-\eta\;\nabla_{\theta}\mathcal{L}(\theta)$

where, 
<ul>
    <li>$\theta\in(w,b)$</li>
    <li>$\eta$ : learning rate</li>
    <li>$\mathcal{L}$ : cost function</li>
</ul>

It needs optimization for many epochs until convergence which depends on learning rate and number of epochs.

SGD scales better for huge datasets or online learning, where OLS is too expensive.

LinearRegression = "one-shot exact solve" (analytical/numerical linear algebra)

SGDRegressor = "trial-and-error until close enough" (optimization-based, scalable)

Mathematically, from bias-variance decomposition, $\text{Error}_{\text{total}} = \text{Bias}^2 + \text{Variance} + \text{Irreducible Noise}$. <i>Need to say accurately !!!</i>

Here the $\text{Bias}$ is not the intercept, rather it is an error from incorrect assumptions of model that is associated with training dataset and $\text{Variance}$ is defined for test dataset. Our motive should be always for low $\text{Bias}$ and low $\text{Variance}$ for generalized model.

Sometimes, training data poorly performs with model. Then it is called Underfitting. Here, $\text{Bias}$ stays high.

Overfitting occurs when the model learns not only the underlying relationship between the features $X$ and the target $Y$ but also the random noise present in the training data. This makes the model perform very well on the training data but poorly on test data (new data). In other words, training error is very low ($\approx 0$) but test error is high. It corresponds to very low $\text{Bias}$ but very high $\text{Variance}$.

It happens when, 
<ul>
    <li>The number of features is large relative to the number of samples.</li>
    <li>The estimated coefficients have high variance, leading to a model that fits even small fluctuations in the training set.</li>
</ul>


Regularization is a technique used in machine learning to prevent overfitting by adding a penalty term to the loss function. There are mainly two types of regularization techniques depending on the types of coefficients -
<ol>
    <li>Ridge</li>
    <li>Lasso</li>
    <li>Elastic-Net</li>
</ol>

In <b>Lasso (L1)</b>, the penalty term is proportional to the absolute value of coefficients.
$$\mathcal{L}(b,w_0,w_1,...,w_n) = \frac{1}{2N} \sum_{i=1}^{N}\;(y_i-\hat{y}_i)^2 + \lambda \sum_{j=0}^{n}\;|w_j|$$ 
where, $\lambda$ ($\geq0$) is regularization strength. This penalty causes some coefficients to shrink exactly to zero because the penalty term is an equation of octahedron (in higher dimension). This help to evaluate feature selection.

In <b>Ridge (L2)</b>, the penalty term is proportional to the square of coefficients.
$$\mathcal{L}(b,w_0,w_1,...,w_n) = \frac{1}{2N} \sum_{i=1}^{N}\;(y_i-\hat{y}_i)^2 + \lambda \sum_{j=0}^{n}\;|w_j|^2$$ If $\lambda$ is larger, coefficients shrink closer to zero but never become exactly zero because the penalty term is an equation of hypersphere (in higher dimension).

In <b>Elastic-Net</b>, both $\text{Lasso}$ and $\text{Ridge}$ are consider as the penalty term.

In regression problem statement, we can measure how well the regression predictions approximate the real data points compared to a mean-based model. The quantity is called <b>$\text{R}^2$ score</b>, defined by, 
$$\text{R}^2 = 1-\frac{\sum{(y_i-\hat{y_i})^2}}{\sum{(y_i-\bar{y})^2}}$$
where, $y_i$ = Actual/True value, $\hat{y}$ = Predicted value, $\bar{y_i}$ = Mean of the true value

| **$R^2$ Score** | **Interpretation** |
| :-----: | :-----: |
| $=1$ | Perfect Prediction |
| $=0$ | Model is no better than predicting the mean |
| $<0$ | Model is worse than just predicting the mean |

In other words, this quantity explains goodness-of-fit (or variance) of the data to the model.

##### <span style="color:#BB1C75">Standardization</span>

It is a feature scaling technique where the values of each feature are transformed so that they have the properties of a standard normal distribution i.e., zero mean and unit standard deviation. If $x$ is the old feature with mean $\mu$ and standard deviation $\sigma$ then the new feature $x^{'}$ can be written as, 

$$x^{'}=\frac{x-\mu}{\sigma}$$

This technique must be applied for distance-based and gradient-based algorithm such as - Linear Regression, Logistic Regression, SVM, K-Nearest Neighbors, Clustering, Principal Component Analysis and as well as Neural Network.

Not necessary for tree-based + rule-based algorithm such as - Naive Bayes, Decision Trees, Random Forests, Gradient Boosting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import corner
from tqdm import tqdm
```

##### <span style="color:#BB1C75">Single-featured</span>

Here, we generate a noisy data with an injected straight line of $\tan{\theta}=8$ and intercept of 3

```python
np.random.seed(42)

n_samples = pow(2,10)*8
n_feature = 1

w1 = 8
b = 3
x1 = np.linspace(1, 5, n_samples)

def y_true(w1,b,x1):
    return (w1*x1)+b

y = y_true(w1,b,x1) + np.random.randn(n_samples)
```

Now we have to seperatly consider train data and test data from the target data set. We can use <i>sklearn.model_selection.train_test_split</i>

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x1, y, test_size=0.2, random_state=42)

df_train = pd.DataFrame({'X_train':x_train,'Y_train':y_train})
df_test = pd.DataFrame({'X_test': x_test})
```

Call the Ordinary least squares Linear Regression from sklearn.linear_model.

```python
reg_OLS = sklearn.linear_model.LinearRegression()
```

Here, <i>fit</i> helps to fit the model parameters (weight and bias) using training data which contains the coefficient and the intercept. Later we predict the output using the test dataset.

```python
reg_OLS.fit(df_train.X_train.values.reshape(len(x_train),n_feature), df_train.Y_train.values.reshape(len(x_train),))
```

Estimated weights and bias/intercept term

```python
print('Weights :',reg_OLS.coef_)
print('Intercept :',reg_OLS.intercept_)
```

Predict output with the test dataset

```python
y_pred = reg_OLS.predict(df_test.X_test.values.reshape(len(x_test),n_feature))
```

```python
r2 = sklearn.metrics.r2_score(y_test, y_pred)
r2
```

```python
plt.scatter(y_test, y_pred, alpha=0.3, s=3)
plt.plot(y_test, y_test, color='red', ls='--', lw=0.8)
plt.xlabel('True', labelpad=20)
plt.ylabel('Predict', labelpad=20)
```

##### <span style="color:#BB1C75">SGD</span>

We minimize the emperical loss using stochastic process.

Let's create a random dataset for regression problem using sklearn.datasets with Gaussian noise of unit standard deviation for single-valued output.

```python
X,Y,W = sklearn.datasets.make_regression(n_samples=10000, n_features=1, 
                                 n_informative=1, n_targets=1, bias=0, 
                                 effective_rank=None, tail_strength=0.5, noise=1, 
                                 shuffle=True, coef=True, random_state=42)


x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.35, random_state=42)


reg_SGD = sklearn.linear_model.SGDRegressor(fit_intercept=False)
reg_SGD.fit(x_train, y_train)

y_pred = reg_SGD.predict(x_test)
```

```python
plt.scatter(y_test, y_pred, alpha=0.3, s=3)
plt.plot(y_test, y_test, color='red', ls='--', lw=0.8)
plt.xlabel('True', labelpad=20)
plt.ylabel('Predict', labelpad=20)
```

##### <span style="color:#BB1C75">Cross-Validation</span>

If we train the data only once, we might get a biased results (lucky/unlucky split). Through cross-validation, we split our training data and train each sets with the same model. This gives a more reliable estimate of generalization performance of our model.

```python
score = sklearn.model_selection.cross_val_score(reg_SGD, x_train, y_train, cv=10, scoring='neg_mean_squared_error')
score
```

This score signifies the model performs quite well. We can set the 'scoring' with our preference.

##### <span style="color:#BB1C75">Hyperparameter Tuning</span> :

It is the process of finding the best set of hyperparameters that gives the best performance on test data. Since hyperparameters are not learned during training, we try different values and use a performance metric to decide which is best. Although SGDRegressor doesn't have much hyperparameters but we can try with learning rate, regularization, maximum iteration.

We use 'GridSearchCV' to search over specified parameter values for an estimator. Those parameters of the estimator used to optimize by cross-validated grid search over a parameter grid.

```python
parameters = {'penalty':['l1','l2','elasticnet'], 'max_iter':[1000,100,5000,10000], 'eta0':[1e-1,1e-2,1e-3]}

reg_SGD_hyp = sklearn.model_selection.GridSearchCV(reg_SGD, param_grid=parameters, scoring='neg_mean_squared_error', cv=20)
reg_SGD_hyp.fit(x_train, y_train)
```

The best selected parameter setting for trainning the data can be seen using 'best_params_'

```python
reg_SGD_hyp.best_params_
```

```python
y_pred = reg_SGD_hyp.predict(x_test)
print(sklearn.metrics.r2_score(y_test, y_pred))
```

In terms of increasing the score, we must play with other/more parameters

As the process is stochastic, each iteration of .fit() produce different values of coefficients and intercept. Here, comes the optimization. We can obtain the parameter samples by an iterative process.

```python
for itr in range(5):
    reg_SGD.fit(x_train, y_train)
    print(reg_SGD.coef_)
```

##### <span style="color:#BB1C75">Double-featured</span>

Let's create a random dataset for regression problem with double features.

```python
X,Y,W = sklearn.datasets.make_regression(n_samples=1000, n_features=2, 
                                 n_informative=2, n_targets=1, bias=100, 
                                 effective_rank=None, tail_strength=0.5, noise=1, 
                                 shuffle=True, coef=True, random_state=42)


x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.2, random_state=None)


loss,bias,weight_1,weight_2 = [],[],[],[]

for itr in tqdm(range(10000)):
    reg_SGD = sklearn.linear_model.SGDRegressor(learning_rate='constant')
    reg_SGD.fit(x_train, y_train)
    weight_1.append(reg_SGD.coef_[0])
    weight_2.append(reg_SGD.coef_[1])
    bias.append(reg_SGD.intercept_[0])
    y_pred = reg_SGD.predict(x_test)
    loss.append(sklearn.metrics.mean_squared_error(y_test, y_pred))

fig = plt.figure(figsize=[7,5])

corner.hist2d(np.array(weight_1), np.array(weight_2), levels=[0.5,0.9], color='fuchsia', plot_datapoints=False, plot_density=False)
f = plt.scatter(weight_1, weight_2, c=loss, s=1, cmap='Blues')
plt.scatter(W[0], W[1], marker='*', s=50, color='lime', label='Injected')
plt.xlabel(r'$\text{w}_1$', labelpad=20)
plt.ylabel(r'$\text{w}_2$', labelpad=20)
plt.colorbar(f, label='Loss (MSE)')
plt.legend()
```

### <span style="color:#BB1C75">Logistic Regression</span>

Logistic regression is used for classification problem statement (Despite its name, it is a classification algorithm, not regression) where the outputs are the probability distribution. Here, I will discuss about binary classification where our decision consist of two classes (Let's say True and False).

We generally use <i>sigmoid</i> function as the model, $$P(z=\text{True}\;|\;x_1,x_2,...,x_n) = \sigma(z) = \frac{1}{1+e^{-z}}\;\;,\;\text{where}\;z=b+\beta_1x_1+\beta_2x_2+...+\beta_nx_n$$

Here $\sigma(z)$ is called <i>Sigmoid Activation Function</i> whose value belongs from $0$ to $1$. Decision is based on a threshold of the probability (mainly $0.5$). 

| **z** | **$\sigma(z)$** | **Decision (T/F)** |
| :-----: | :-----: | :-----: |
| $>0$ | $>0.5$ | T |
| $<0$ | $<0.5$ | F |

The cost function, $$\mathcal{L}(b,w_0,w_1,...,w_n) = \frac{1}{N} \sum_{i=1}^{N}\;[y_i\;\text{log}_{e}\{{\hat{p}(x_i)}\}+(1-y_i)\;\text{log}_{e}\{1-{\hat{p}(x_i)}\}]$$

Where, $\;\hat{p}(x_i)$ = <span style="font-size:20px;">$\frac{1}{1+e^{-(b+\beta x_i)}}$</span> 

For convergence, we need to minimize the cost function obeying, $\theta =: \theta-\eta\;\nabla_{\theta}\mathcal{L}(\theta)$

In classification problem statement, we can measure the fraction of correctly predicted classes out of the total number of predictions. The quantity is called <b>accuracy score</b>, defined by, 
$$\text{Accuracy} = \frac{\text{Number of correct predictions​}}{\text{Total number of predictions}}$$
It ranges from $0$ to $1$.

```python
X,Y = sklearn.datasets.make_classification(n_samples=1000, n_features=1, n_classes=2, 
                                           n_informative=1, n_redundant=0, n_repeated=0, 
                                           n_clusters_per_class=1, random_state=42)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.45, random_state=None)

cls = sklearn.linear_model.LogisticRegression()
cls.fit(x_train, y_train)
y_pred = cls.predict(x_test)
```

To calculate the probability of each classes for test data points, we can use <i>predict_proba()</i>. The return array is a shape of (n_sample, n_class). Each column quantifies classes for test data and each row defines the probability of both classes. By summing each row, we get the total probability for each classes (i.e. $1$).

```python
y_prob = cls.predict_proba(x_test)
```

```python
plt.scatter(x_test, y_test, s=10, alpha=0.5, label='Test dataset')
plt.scatter(x_test, y_prob[:,0], s=10, alpha=0.5, label=r'P$\;(y=0)$')
plt.scatter(x_test, y_prob[:,1], s=10, alpha=0.5, label=r'P$\;(y=1)$')
plt.xlabel('Y', labelpad=20)
plt.ylabel(r'$\sigma\;(Y)$', labelpad=20)
plt.legend()
```

```python
X,Y = sklearn.datasets.make_classification(n_samples=1000, n_features=2, n_classes=2, 
                                           n_informative=2, n_redundant=0, n_repeated=0, 
                                           n_clusters_per_class=1, random_state=42)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.45, random_state=None)

cls = sklearn.linear_model.LogisticRegression()
cls.fit(x_train, y_train)
y_pred = cls.predict(x_test)

w = cls.coef_
b = cls.intercept_
x_2 = -(b/w[0][1] + (w[0][0]/w[0][1])*np.sort(x_test[:,0]))

plt.scatter(x_train[:,0], x_train[:,1], c=y_train, s=10, cmap='winter')
plt.plot(np.sort(x_test[:,0]), x_2, color='red', ls='--', label=r'z$=0$ ; P(z)$=0.5$')
plt.xlabel(r'$x_1$', labelpad=20)
plt.ylabel(r'$x_2$', labelpad=20)
plt.legend()
plt.colorbar()
```

##### <span style="color:#BB1C75">Performance Matrices</span>

Performance matrices summarize how well the model performs. There are many quantitative measures derived from model predictions -

<ul>
    <li>Accuracy</li>
    <li>Precision</li>
    <li>Recall</li>
    <li>F-$\beta$ score</li>
</ul>

Confusion matrix is a tabular representation of prediction outcomes that shows the detailed distribution of correct and incorrect predictions. Consider a binary classification problem statement with output (True/False). This can be calculated using $\href{https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html#sklearn.metrics.confusion_matrix}{\underline{\text{sklearn.metrics.confusion\_matrix}}}$


|  | **True($1$)** | **False($0$)** |
| :-----: | :-----: | :-----: |
| **True($1$)** | TP | FP |
| **False($0$)** | FN | TN |

Here, T = True, F = False, TP = True Positive, TN = True Negative and so on. 

Rows and Coulmns are associated with Predicted and True values repectively. We need to fill this table by putting the total number of desired combination.

    
**Accuracy**$\;\;\;$=$\;\;\;$<span style="font-size:20px;">$\frac{\text{TP}+\text{TN}}{\text{TP}+\text{FP}+\text{FN}+\text{TN}}$</span>

If our train dataset is <i>Imbalanced</i> i.e., number of True and number of False are highly differ then we can't use <i>Accuracy</i>. Instead of this, we need to use <i>Precesion</i> and <i>Recall</i>

**Precesion**$\;\;\;$=$\;\;\;$<span style="font-size:20px;">$\frac{\text{TP}}{\text{TP}+\text{FP}}$</span>$\;\;\;\;\;\;$, $\;\;\;\;\;\;$**Recall**$\;\;\;$=$\;\;\;$<span style="font-size:20px;">$\frac{\text{TP}}{\text{TP}+\text{FN}}$</span>$\;\;\;\;\;\;$, $\;\;\;\;\;\;$**F-$\beta$ score**$\;\;\;$=$\;\;\;$<span style="font-size:20px;">$\frac{(1+\beta^2)\;*\;\text{Precesion}\;*\;\text{Recall}}{\beta^2\;(\text{Precesion}+\text{Recall})}$</span>

So, in which case we need to use them ? Let's take an example of <i>spam classification</i>.

Mail = spam, Model = spam ---> Good ; Mail = not spam, Model = not spam ---> Good ; Mail = spam, Model = not spam ---> Fine ! ; Mail = not spam, Model = spam ---> Blunder !!!

Now, we will see the confusion matrix.

|  | **Spam** | **Not Spam** |
| :-----: | :-----: | :-----: |
| **Spam** | Good | Blunder !!! |
| **Not Spam** | Fine ! | Good |

This shows FP is the <i>blunder</i>. Hereby, we will use <i>Precesion</i> where we need to reduce the FP by tunning hyperparameters.

For, <i>Recall</i>, we will reduce the FN by tunning hyperparameters. E.g.,

Person = AIDS, Model = AIDS ---> Good ; Person = not AIDS, Model = not AIDS ---> Good ; Person = AIDS, Model = not AIDS ---> Blunder !!! ; Person = not AIDS, Model = AIDS ---> Need to check by that person !

|  | **AIDS** | **Not AIDS** |
| :-----: | :-----: | :-----: |
| **AIDS** | Good | Need to check ! |
| **Not AIDS** | Blunder !!! | Good |


Another matrix is called <i>F-$\beta$ score<i>.

<ul>
    <li>If both FP and FN is important then $\beta=1$. This is called <i>Hermonic Mean</i>. For example, crash of the stock market, where both customer and company must have their own priorities.</li>
    <li>When FP>FN ; FP is more important than FN, $\beta=0.5$</li>
    <li>When FN>FP ; FN is more important than FP, $\beta=2$</li>
</ul>

These quantities can be calculated using $\href{https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html#sklearn.metrics.classification_report}{\underline{\text{sklearn.metrics.classification\_report}}}$

##### <span style="color:#BB1C75">Naive Bayes' Classifier</span>

Bayes' theorem states that, $\text{P}(A|B) = \large \frac{\text{P}(B|A)\;\text{P}(A)}{\text{P}(B)}$

Here, we will apply Bayes' theorem by assumming the conditional independent of features given the classes. The posterior probability of a particular class is given by , 

$$\;\text{P}(y|x_1,x_2,...,x_n) = \frac{\text{P}(x_1,x_2,...,x_n|y)\;\text{P}(y)}{\text{P}(x_1,x_2,...,x_n)}$$

where, $y$ is the class (for binary classification, $y\in\{0,1\}$) and the set of feature is $\{x_1,x_2,...,x_n\}$

We can decompose the likelihood as well as the denominator along each feature i.e., $\text{P}(x_1,x_2,...,x_n) = \text{P}(x_1)\;\cdot\;\text{P}(x_2)\;\cdot\cdot\cdot\;\text{P}(x_n)$. Also, the denominator is a constant terms, so we can write, 

$$\;\text{P}(y|\{x\})\;\propto\;\text{P}(y)\;\prod_{i=1}^{n}\;\text{P}(x_i|y)$$

##### <span style="color:#BB1C75">k-Nearest Neighbour</span>

KNN is used as both regression and classification problem statement. For both problem statements, the main goal is to find a suitable output according to the trained model.

Suppose, we train a model for multi-class catagories. Using KNN, we can check that in which catagory, the test datapoints belong.

The procedure is so simple. Depending on the value of $k$, it create the nearest points of a test point according by measuring the distance between the test data and the nearest data points. Then, the class is decided by the maximum number of neighbour points of that class. 

Now, distance can be calculated using different types of approach. One of the common is euclidean distance (see $\href{https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.distance_metrics.html#distance-metrics}{\underline{\text{here}}}$ for other metrics). The value of $k$ vary with different datasets and estimated using hyperparameter tunning. For regression problem statement, the process is same but the output will be average of all the distances from its nearest neighbours.

This KNN classification has some limitations - 
<ul>
    <li>Lazy with huge dataset</li>
    <li>Sensitive to missing values and outliers</li>
</ul>

```python
X,Y = sklearn.datasets.make_classification(n_samples=2000, n_features=2, n_classes=2, 
                                           n_informative=2, n_redundant=0, n_repeated=0, 
                                           n_clusters_per_class=1, random_state=42)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.4, random_state=None)

knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
```

### <span style="color:#BB1C75">Support Vector Machine (SVM)</span>

A Support Vector Machine is a supervised machine learning algorithm used for classification (SVC) and regression (SVR) statement problem. It works by finding the best separating boundary (so called a hyperplane) that has maximized seperating distance of two extra equi-distanced hyperplanes along with decision plane for both classes, considering nearest points. These two hyperplanes are called <i>Marginal plane</i>. The marginal planes are drawn for the closest points (from training dataset) from each class that lie exactly on the marginal planes. These are called the <i>support vectors</i>. The entire marginal boundary is determined only by these support vectors. If there is no outliers (or points of a class stay at other classes), then the hyperplane is called <i>Hard-Marginal Plane</i>. Otherwise, it is called <i>Soft-Marginal Plane</i>. Ideally, we mostly have soft-marginal plane i.e., no training points should lie inside the margin. 

####

##### <span style="color:#BB1C75">SVC</span>

Geometric Intuition of SVC is quite like, suppose you have two classes of points and many hyperplanes may separate them perfectly. But the one with the largest margin is the most stable, because it relies only on the most critical points (support vectors), not the rest. This leads to a unique optimal solution instead of infinitely many separating hyperplanes.

<br>

  <img src="SVM.png" width="350"/>
</div>

<br>

The distance of the best fitted hyperplane (H0) from the center : $$\frac{\vec{\text{w}}\cdot\vec{\text{x}}}{||\vec{\text{w}}||}=\mathcal{C}\;(\text{constant})$$

<br>

| **Decision** | **Equation** |
| :-----: | :-----: |
| Best Fit (H0) | $\vec{\text{w}}\cdot\vec{\text{x}}+b=0$ |
| H1 | $\vec{\text{w}}\cdot\vec{\text{x}}+b=+k$ |
| H2 | $\vec{\text{w}}\cdot\vec{\text{x}}+b=-k$ |

<br>

For Mathematical convenient, we often consider $k=1$


SVM is a maximum margin classifier. The reason of maximizing the distance between marginal planes are ---

<ul>
    <li>If the margin is large, small variations or noise in the data won’t change the classification easily</li>
    <li>A small margin means the boundary is very close to the data, even a tiny perturbation could cause misclassification</li>
    <li>Maximizing the margin reduces the model's tendency to overfit the training data</li>
    <li>A larger margin, less dependent on specific training points, generalizes better to unseen data</li>
</ul>

<br>

For two points of different points A and B, the distance between them can be written as : $(\vec{\text{x}}_{+}-\vec{\text{x}}_{-})\cdot\hat{\text{w}}$

$$(\vec{\text{x}}_{+}-\vec{\text{x}}_{-})\cdot\hat{\text{w}} = \frac{(\vec{\text{x}}_{+}\cdot\hat{\text{w}})-(\vec{\text{x}}_{-}\cdot\hat{\text{w}})}{||\vec{\text{w}}||} = \frac{(1-b)-(-1-b)}{||\vec{\text{w}}||} = \frac{2}{||\vec{\text{w}}||}$$

<br>

We need to maximize this term in order to minimize the $||\vec{\text{w}}||$ or $\frac{1}{2}(\vec{\text{w}}\cdot\vec{\text{w}})$

<br>

| **Decision** | **Constraint** |
| :-----: | :-----: |
| H1 $(y=+1)$ | $(\vec{\text{w}}\cdot\vec{\text{x}_{i}}+b)\geq1$ |
| H2 $(y=-1)$ | $(\vec{\text{w}}\cdot\vec{\text{x}_{i}}+b)\leq-1$ |

<br>

In general, we need to optimize $\frac{1}{2}(\vec{\text{w}}\cdot\vec{\text{w}})$, subject to $y(\vec{\text{w}}\cdot\vec{\text{x}_{i}}+b)\geq1$.


In Mathematical optimization, the method of Lagrange multipliers is a strategy for finding the local maxima and minima of a function subject to equality constraints.

$$L = \frac{1}{2}(\vec{\text{w}}\cdot\vec{\text{w}})-\sum_{i}\alpha_{i}\;[y_{i}(\vec{\text{w}}\cdot\vec{\text{x}_{i}}+b)-1]$$

<br>

$\frac{\partial{\text{L}}}{\partial{\text{w}}} = \vec{\text{w}}-\sum_{i}\alpha_{i}y_{i}\vec{\text{x}_{i}}=0\;\Rightarrow\;\vec{\text{w}}=\sum_{i}\alpha_{i}y_{i}\vec{\text{x}_{i}}$

$\frac{\partial{\text{L}}}{\partial{\text{b}}}=\sum_{i}\alpha_{i}y_{i}=0$

<br>

It tells us that $\vec{\text{w}}$ can be obtained with a linear sum of samples. At the end of optimization, lot of $\alpha_i$ is going to be zero and the non-zero $\alpha_i$ correspond to the samples in the gutter, which are the support vectors.

We put $\vec{\text{w}}$ back to the Lagrangian equation,

$\begin{eqnarray}
\text{L} &=& \frac{1}{2}(\sum_{i}\alpha_{i}y_{i}\vec{\text{x}_{i}})(\sum_{j}\alpha_{j}y_{j}\vec{\text{x}_{j}})-(\sum_{i}\alpha_{i}y_{i}\vec{\text{x}_{i}})(\sum_{j}\alpha_{j}y_{j}\vec{\text{x}_{j}})+\sum_{i}\alpha_{i}y_{i}b-\sum_{i}\alpha_{i} \\
\\
&=& \sum_{i}\alpha_{i} - \frac{1}{2}\sum_{i}\sum_{j}\alpha_{i}\alpha_{j}y_{i}y_{j}(\vec{\text{x}_{i}}\cdot\vec{\text{x}_{j}})
\end{eqnarray}$


Optimization depends only on the dot product of pairs of samples. The above formalism is for Hard margin SVM.

For soft margin SVM, the constraints have to be satisfied : $y_i(\vec{\text{w}}\cdot\vec{\text{x}_{i}}+b)\geq1-\xi_{i}$ ; $\;\;\;(i=1,2,...,n)$. Here, we could apply L1 regularization to penalize the large value of $\xi$. Then, the cost function can be written as,

$$\mathcal{L}(w,b,\xi) = \frac{\vec{\text{w}}\cdot\vec{\text{w}}}{2} + \text{C}\;\sum_{i=1}^{n}\;\xi_i\;\;\;\;|\;\;\;\;y_i(w^{T}x_i+b)\geq1-\xi_i$$

where,
<ul>
    <li>$\text{C}$ : The number of points that can be avoided for misclassification</li>
    <li>$\xi_i (\geq0)$ : Distance of misclassified points from marginal planes</li>
</ul>

These two quantity are the hyperparameters. The second term in the cost function is called <i>Hinge Loss</i>. A smaller $\text{C}$ emphasis the importance of $\xi$ and a larger $\text{C}$ diminishes the importance of $\xi$. When $\text{C}$ becomes $+\infty$, the optimization results behave as hard margin SVC. Again, smaller $\text{C}$ indicates wider margin (the better SVC generalizes, less sensitive to noise) and large $\text{C}$ indicates narrower margin.


Again, our aim should be always to maximize the distance between the marginal hyperplanes. Inversely, it could be done by minimizing the cost function for $w,b$ and $\xi$.

####

##### <span style="color:#BB1C75">SVR</span>

For regression problem statement, if the best fit line is defined as, $\;y_i=(w^Tx_i+b)$ then we can write the marginal hyperplanes as $\;y_{+}=(w^Tx_i+b+\epsilon)$ and $\;y_{-}=(w^Tx_i+b-\epsilon)$, where $\epsilon$ is the distance between best fit line and hyperplanes along the output for both .


The cost function is wriiten as -
$$\mathcal{L}(w,b,\xi) = \frac{w^{T}w}{2} + \text{C}\;\sum_{i=1}^{n}\;\xi_i\;\;\;\;|\;\;\;\;|y_{i}-\hat{y_i}|\leq\epsilon+\xi_i$$

where,
<ul>
    <li>$\text{C}$ : The number of points which can be considered as the outliers or the points those are outside of the hyperplanes</li>
    <li>$\xi_i (\geq0)$ : Distance of these points from marginal planes of respective side</li>
</ul>

Similar to SVC, the aim should be always to maximize the distance between the marginal hyperplanes. Inversely, it could be done by minimizing the cost function for $w,b$ and $\xi$.

If our dataset is not linearly separable then we need to use SVM kernels.

####

##### <span style="color:#BB1C75">SVM Kernel</span>

Soft margin SVM can handle the non-linearly separable data. Here the kernel trick came !!

The Lagrange equation we got : $\;\;\;\text{L} = \sum_{i}\alpha_{i} - \frac{1}{2}\sum_{i}\sum_{j}\alpha_{i}\alpha_{j}y_{i}y_{j}\;\mathcal{K}(\vec{\text{x}_{i}},\vec{\text{x}_{j}})$

Where, $\mathcal{K}(\vec{\text{x}_{i}},\vec{\text{x}_{j}})$ is the kernel function that is used to transform the input data into a higher dimensional feature space. Kernel helps us to make a trasformation of the coordinates so that we can seperate the non-linear datapoints of each classes. There are different types of kernels --

<ul>
    <li>Linear Kernel : $\mathcal{K}(\vec{\text{x}_{i}},\vec{\text{x}_{j}})=(\vec{\text{x}_{i}}\cdot\vec{\text{x}_{j}})$ ; No transformation, works well in the original feature space. This is useful when the data is linearly separable.</li><br>
    <li>Polynomial Kernel : $\mathcal{K}(\vec{\text{x}_{i}},\vec{\text{x}_{j}})=(\vec{\text{x}_{i}}\cdot\vec{\text{x}_{j}}+\text{c})^{d}$ ; $\text{c}$ is a constant and $d$ is degree of freedom. A higher $d$ value could make the decision boundary too complex and might result in overfitting.</li><br>
    <li>Radial Basis Function (RBF) / Gaussian Kernel : $\mathcal{K}(\vec{\text{x}_{i}},\vec{\text{x}_{j}})=e^{-\gamma||\vec{\text{x}_{i}}-\vec{\text{x}_{j}}||^{2}}$ ; where $\gamma (>0)$ is a constant. This is a very popular kernel that maps data into infinite-dimensional space. It Works well in most practical cases when data is not linearly separable. A small value of $\gamma$ behaves like linear SVM without kernels and large $\gamma$ will make the model heavily impacted by the support vectors and may result overfitting.</li>
</ul>

By tuning hyperparameters, we can check which kernel better distinguish our dataset.

######

```python
X,Y = sklearn.datasets.make_classification(n_samples=1000, n_features=2, n_classes=2, 
                                           n_informative=2, n_redundant=0, n_repeated=0, 
                                           n_clusters_per_class=1, class_sep=2, random_state=42)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.35, random_state=None)

svc = sklearn.svm.SVC(C=5, kernel='linear', degree=2)
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)

print('Accuracy :', sklearn.metrics.accuracy_score(y_test, y_pred))
```

##### <span style="color:#BB1C75">Perceptron</span>

A perceptron is the basic building block of a neural network. It represents a single neuron that could take several inputs (features), weights, and bias.

<br>

Neuron is divided into two steps --
<ul>
    <li><b>Step:1</b> $\;\longrightarrow\;$ The input values are multiplied by their respective weights by adding a bias. $\sum_{i=1}^{n}\;w_{i}x_{i}+b_{1}$</li><br>
    <li><b>Step:2</b> $\;\longrightarrow\;$ We apply an actication function to get the output for that hidden layer. Activation function tells the neurons whether to activate or not. For example, binary classification : $\sigma(\sum_{i=1}^{n}\;w_{i}x_{i}+b_{1})$. This act as the output of that neuron and the input of the output layer.</li>
</ul>


<i>Output layer :</i> Same as hidden layer, it takes inputs from the previous hidden layer output, multiply weights and add bias then apply an activation function. For example, binary classification : $\;\hat{y} = \sigma(\{\sigma(\sum_{i=1}^{n}\;w_{i}x_{i}+b_{1})\}w_{0}+b_{2})$.


Now, There is two important procedure to train the neural network --

<ul>
    <li><b>Forward Propagation :</b> Took the inputs, multiply with weights by adding a bias and apply activation function. Then predict an ouput. But our true value is $y$. So, cost function $\propto (y-\hat{y})$. We have to minimize the cost function !</li><br>
    <li><b>Backward Propagation :</b> In order to minimize cost function, we need to do update the weights and bias and perform forward propagation. This updation must continue until the difference between predicted and true value become minimum. This is called Backward Propagation. It is done by " Optimizer ".</li>
</ul>

<br>

<div align="center">
  <img src="Train_NN.png" width="500"/>
</div>

<br><br>

Following the above figure, we train a neural network.

The updation of weights and bias will be followed by, $$\theta_{\text{new}} = \theta_{\text{old}}-\eta\;\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{old}}}}$$

where, $\mathcal{L}$ is the cost function and $\eta$ is the learning rate.

Now, we need to evaluate the term $\begin{eqnarray}\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{old}}}}\end{eqnarray}$ for each weights and bias. So, we will use chain rule of differentiation. First let's consider the output of neuron is $\mathcal{O}_1$ and the output of the output layer is $\mathcal{O}_2$ which is actually $\hat{y}$. We can write them as, $\mathcal{O}_1 = \sigma(\sum_{i=1}^{n}\;w_{i}x_{i}+b_{1})$ and $\mathcal{O}_2 = \sigma(\{\sigma(\mathcal{O}_1)\}w_{0}+b_{2})$.

<br>

$\begin{eqnarray}
\frac{\partial{\mathcal{L}}}{\partial{w_{0_\text{old}}}} &=& \frac{\partial{\mathcal{L}}}{\partial{\mathcal{O}_2}}\ast\frac{\partial{\mathcal{O}_2}}{\partial{w_{0_\text{old}}}}\;\;;\;\;\frac{\partial{\mathcal{L}}}{\partial{w_{1_\text{old}}}} &=& \frac{\partial{\mathcal{L}}}{\partial{\mathcal{O}_2}}\ast\frac{\partial{\mathcal{O}_2}}{\partial{\mathcal{O}_1}}\ast\frac{\partial{\mathcal{O}_1}}{\partial{w_{1_\text{old}}}} \\
\end{eqnarray}$

<br>

Similarly,

$\begin{eqnarray}
\frac{\partial{\mathcal{L}}}{\partial{w_{n_\text{old}}}} &=& \frac{\partial{\mathcal{L}}}{\partial{\mathcal{O}_2}}\ast\frac{\partial{\mathcal{O}_2}}{\partial{\mathcal{O}_1}}\ast\frac{\partial{\mathcal{O}_1}}{\partial{w_{n_\text{old}}}}
\end{eqnarray}\;\;\;$;$\;\;\;$here, $\mathcal{O}_1$ and $\mathcal{O}_2$ are connected by $w_{0}$. So, we can write $\begin{eqnarray}
\frac{\partial{\mathcal{O}_2}}{\partial{\mathcal{O}_1}} &=& \frac{\partial{\mathcal{O}_2}}{\partial{w_{0}}}\ast\frac{\partial{w_{0}}}{\partial{\mathcal{O}_1}}\end{eqnarray}$

<br><br>

  <img src="NN_exmp.png" width="600"/>
</div>

<br><br>

For this figure, $\;\;\;\;\;\begin{eqnarray}
\frac{\partial{\mathcal{L}}}{\partial{w_{1}}} &=& \frac{\partial{\mathcal{L}}}{\partial{\mathcal{O}_3}}\ast\frac{\partial{\mathcal{O}_3}}{\partial{\mathcal{O}_2}}\ast\left[\left(\frac{\partial{\mathcal{O}_2}}{\partial{\mathcal{O}_{11}}}\ast\frac{\partial{\mathcal{O}_{11}}}{\partial{\mathcal{O}_1}}\ast\frac{\partial{\mathcal{O}_1}}{\partial{w_{1}}}\right) + \left(\frac{\partial{\mathcal{O}_2}}{\partial{\mathcal{O}_{22}}}\ast\frac{\partial{\mathcal{O}_{22}}}{\partial{\mathcal{O}_1}}\ast\frac{\partial{\mathcal{O}_1}}{\partial{w_{1}}}\right)\right]
\end{eqnarray}$

<br>

Optimizer does this chain rule and update the cost function.

###

##### <span style="color:#BB1C75">Vanishing Gradient Problem</span>

Suppose, we have a deep network with lot of neurons. When we update the weights and bias, we need to apply the chain rule. For binary classification, we mainly use <i>sigmoid function</i> to calculate the cost function. Each output of neuron is a sigmoid function, that means in chain rule, we naturally calculate the derivative of the sigmoid function. Now, the derivation of a sigmoid function is Gaussian in nature whose value is bounded between $0$ to $0.25$. So, in the chain rule, every differentiation term has maximum value $0.25$. But multiplying those terms for many times gives a very very small value. Even the learning rate, $\eta\sim 0.001\;$or$\;0.01$. Overall, the second term of the updation formula get nearly zero. Then $w_\text{new}\approx w_\text{old}$, that means the parameter will stuck forever. This is called <i>Vanishing Gradient Problem</i>.

To overcome this problem, we will use other activation functions.

###

##### <span style="color:#BB1C75">Optimization</span>

First we need to understand what is epoch, batch and iteration ?

<i>Epoch :</i> One forward propagation and one backward propagation of training records/dataset. After one epoch the weights and bias are updated. So, it is important to apply multiple epoches to get rid of underfitting.

<i>Batch :</i> Sometimes, one epoch is too big to feed at once. So, we divide the whole dataset/records into several smaller parts. Each part is called batch. The total number of training records present in a single batch is called the batch size (a hyperparameter).

<i>Iteration :</i> This is the number of batches needed to complete one epoch.

E.g., suppose we have 500 datapoints and the batch size is 50. Then it will take $\begin{eqnarray}\frac{500}{50} &=& 10\end{eqnarray}$ iterations to complete one epoch.

<b>Gradient Decent :</b> We have seen this before. The weights and bias update formula, $\begin{eqnarray}\theta_{\text{new}} = \theta_{\text{old}}-\eta\;\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{old}}}}\end{eqnarray}$. A disadvantage is that when there is a large number of epoch in GD, it becomes resource extensive (Huge RAM needed).

<b>Stochastic Gradient Decent :</b> If we have the $n$ number of records then each epoch consist of $n$ number of iteration. In simplier word, we apply the whole training dataset for each epoch. A disadvantage is the convergence is very slow.

<b>Mini-batch SGD :</b> Suppose, we have $n$ number of records with batch size $p$, then in every epoch consist of $\frac{n}{p}$ number of iterations. An advantage is that it is resource intensive, convergence will be better (than SGD) and the time complexity will improve.

<b>Mini-batch SGD with Momentum :</b> This adds an 'inertia' term to the updates, so the optimizer not only considers the current gradient but also the past update direction. This helps accelerate learning in relevant directions. It is so smoother that it reduce the noise and come to global minima. 

Now, there is a concept of <i>Exponential Weighted Average (EWA)</i>. An exponential weighted average of a sequence $\{x_1,x_2,...,x_n\}$ is defined recursively as, $\begin{eqnarray}\text{V}_\text{t} = \beta\;\text{V}_{\text{t}-1} + (1-\beta)\;\text{X}_\text{t}\end{eqnarray}$

where, 
<ul>
    <li>$\beta \in [0,1]$, is the decay factor (how much 'past' values are remembered). It is a hyperparameter.</li>
    <li>$\text{X}_\text{t}$ is the new observation at step $\text{t}$</li>
    <li>Here $\text{V}_\text{t}$ accumulates the gradient information which act as an inertia term, called velocity.</li>
</ul>


We can write the updation formula in another way, $\begin{eqnarray}\theta_{\text{t}} = \theta_{\text{t}-1}-\eta\;\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{t}-1}}}\end{eqnarray}$.

Here the updation formula is, $\begin{eqnarray}\theta_{\text{t}+1} &=& \theta_{\text{t}}-\text{V}_\text{t}\end{eqnarray}$. Using EWA, we can say, $\begin{eqnarray}\text{V}_\text{t} &=& \beta\;\text{V}_{\text{t}-1} + \eta\;\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{t}}}}\end{eqnarray}$

The advantage is that it reduce the noise and help to make quicker convergence.

<b>AdaGrad (Adaptive Gradient Decent) :</b> Here, our goal is to decrease the learning rate as we move towards the global minima. So, we define the updated learning rate, $$\;\begin{eqnarray}\eta^{'} = \frac{\eta}{\epsilon + \sqrt{\alpha_{\text{t}}}}\end{eqnarray}$$

where, $\;\begin{eqnarray}\alpha_{\text{t}} = \sum_{i=1}^{t}\;\left(\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{t}}}}\right)^2\end{eqnarray}$ and $\epsilon$ is an arbitrary small number to avoid $\alpha_{\text{t}}=0$ condition.

Here, $\alpha_{\text{t}}$ is a summation so it increases with time. As a result, learning rate will decrease over time and that is our goal. But a problem aries for very very deep NN, we could have a large value of $\alpha_{\text{t}}$ that results learning rate so small and $\theta_\text{t}\approx \theta_{\text{t}-1}$ !!

So, we need to make sure that the $\alpha_{\text{t}}$ does not reach very high value. Here, we will apply second moment of EWA, $$\;\begin{eqnarray}\eta^{'} = \frac{\eta}{\epsilon + \sqrt{\mathcal{S}_{\text{t}}}}\end{eqnarray}\;\;\;;\;\;\;\mathcal{S}_{\text{t}} = \;\mathcal{S}_{\text{t}-1} + \left(\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{t}}}}\right)^2$$

<br>

<b>RMSProp :</b> We can apply a decay factor in $\mathcal{S}_{\text{t}}$. $$\mathcal{S}_{\text{t}} = \beta\;\mathcal{S}_{\text{t}-1} + (1-\beta)\;\left(\frac{\partial{\mathcal{L}}}{\partial{\theta_{\text{t}}}}\right)^2$$

<br>

<b>Adam :</b> Here, we combine both SGD with Momentum and Adaptive Gradient Decent with bias correction. 


$$\begin{eqnarray}\hat{\text{V}}_{\text{t}} = \frac{\text{V}_{\text{t}}}{1-\beta_{1}^{\text{t}}}\;\;\;;\;\;\;\hat{\mathcal{S}}_{\text{t}} = \frac{\mathcal{S}_{\text{t}}}{1-\beta_{2}^{\text{t}}}\end{eqnarray}$$

$$\begin{eqnarray}\theta_{\text{t}+1} = \theta_{\text{t}}-\eta^{'}\;\hat{\text{V}}_{\text{t}}\end{eqnarray}\equiv \theta_{\text{t}} - \left(\frac{\eta}{\epsilon + \sqrt{\hat{\mathcal{S}}_{\text{t}}}}\right)\hat{\text{V}}_{\text{t}}$$

This is the best and most popular optimizer.
